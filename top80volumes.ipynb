```python
#importando as bibliotecas

import investpy.stocks
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import datetime as dt
```

```python
# pegando lista de todas as ações do Brasil

Brasil = investpy.stocks.get_stocks(country='Brazil')
display(Brasil)
```

```python
# Adicionando o ".SA" para ficar compatível com os tickers do yahoo finance

tickers = [ticker+'.SA' for ticker in Brasil['symbol']]
print(tickers)
len(tickers)
```

```python
# removendo os ativos com final 11, que podem possuir erros nas cotações

for ticker in tickers:
    if "11.SA" in ticker:
        tickers.remove(ticker)
print(tickers)
len(tickers)
```

```python
#criando uma lista de listas com os 80 ativos mais líquidos de cada janela rolante de 91 dias

listafinal = list()

# criando as janelas rolantes

for c in range(0, 3):
    dias = c*91
    data_final = dt.date.today() - dt.timedelta(days=dias) 
    data_inicial = data_final - dt.timedelta(days=91) 

   # pegando os volumes dos ativos do yahoo finance

    df2 = pd.concat([pdr.get_data_yahoo(tickers,
                                           start=data_inicial,
                                           end=data_final)['Volume']], axis=1)
  
    # calculando a média do volume
    
    meandf= df2.mean()
    
    # Rankeando a média 
    
    ranking = meandf.sort_values(ascending=False)
    
    # Transformando o ranking em dataframe
    
    ranking = pd.DataFrame(ranking)
    
    # Pegando os 80 primeiros
    
    ranking = ranking.head(80)
    
    # Transformando em lista e adicionando na lista de listas
    
    lista = ranking.index.tolist()
    listafinal.append(lista)

print(listafinal)
```
